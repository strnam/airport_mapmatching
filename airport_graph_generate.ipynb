```python
import osmnx as ox
import networkx as nx

%matplotlib inline
```

## Create Changi Aiport Graph from Openstreet Map data

```python
# Load Graph
G = ox.graph_from_place('Singapore Changi Airport', network_type='none',simplify=True,
                        retain_all=True, infrastructure='way["aeroway"]',
                        custom_filter='["aeroway"!~"aerodrome|apron|terminal"]')
ox.plot_graph(G,fig_height=15, fig_width=15)
```

```python
nx.write_yaml(G,'Changi_graph_v1.yaml')
```
